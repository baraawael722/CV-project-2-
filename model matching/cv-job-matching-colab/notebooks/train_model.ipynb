```python
# Import necessary libraries
import pandas as pd
import torch
from src.cv_job_matching_model import CVJobMatcher

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Initialize the CVJobMatcher
matcher = CVJobMatcher(model_name='all-MiniLM-L6-v2')

# Load the data
cvs_file = '/content/data/dataa.csv'
jobs_file = '/content/data/jobs_clean.csv'

# Train the model
best_accuracy = matcher.train(
    cvs_file=cvs_file,
    jobs_file=jobs_file,
    epochs=50,
    batch_size=32,
    learning_rate=0.001
)

# Save the trained model
matcher.save_model('cv_job_matcher_final.pkl')

# Output the best accuracy
print(f"\nðŸŽ‰ Training complete! Best accuracy: {best_accuracy:.2f}%")
```
